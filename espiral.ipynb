# Vizualização de um gráfico 3d em espiral.

### Rozendox

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

ax = plt.axes(projection ='3d')


z = np.linspace(0, 1, 100)
x = z * np.sin(100 * z)
y = z * np.cos(100 * z)
c = x + y
ax.scatter(x, y, z, c = c)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_title('3d')
plt.show()
```
